# Iris Dataset

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("Iris.csv")
df
```

```python
X=df.iloc[:,1:5]
Y=df.iloc[:,-1]
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=42)
```

```python
X_train
```

```python
X_train.shape
```

```python
X_test
```

```python
X_test.shape
```

```python
Y_train
```

```python
Y_train.shape
```

```python
Y_test
```

```python
Y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier()
```

```python
dt.fit(X_train,Y_train)
```

```python
Y_pred=dt.predict(X_test)
```

```python
Y_pred
```

```python
dt.score(X_test,Y_test)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix 
```

```python
accuracy_score(Y_test,Y_pred)
```

```python
precision_score(Y_test, Y_pred, average='micro')
```

```python
recall_score(Y_test, Y_pred, average='micro')
```

```python
f1_score(Y_test, Y_pred, average='micro')
```

```python
confusion_matrix(Y_test, Y_pred)
```

```python
from sklearn import tree
```

```python
clf=tree.DecisionTreeClassifier()
```

```python
clf=clf.fit(X_train,Y_train)
```

```python
tree.plot_tree(clf)
```

# Heart_Disease_Prediction Dataset

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("Heart_Disease_Prediction.csv")
df
```

```python
df.info()
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
X=df.iloc[:,1:14]
Y=df.iloc[:,-1]
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=42)
```

```python
X_train
```

```python
X_train.shape
```

```python
X_test
```

```python
X_test.shape
```

```python
Y_train
```

```python
Y_train.shape
```

```python
Y_test
```

```python
Y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier(criterion="gini")
```

```python
dt.fit(X_train,Y_train)
```

```python
Y_pred=dt.predict(X_test)
```

```python
Y_pred
```

```python
dt.score(X_test,Y_test)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix 
```

```python
accuracy_score(Y_test,Y_pred)
```

```python
precision_score(Y_test, Y_pred, average='micro')
```

```python
recall_score(Y_test, Y_pred, average='micro')
```

```python
f1_score(Y_test, Y_pred, average='micro')
```

```python
confusion_matrix(Y_test, Y_pred)
```

```python
from sklearn import tree
```

```python
clf=tree.DecisionTreeClassifier(criterion="gini",max_depth=3)
```

```python
clf=clf.fit(X_train,Y_train)
```

```python
tree.plot_tree(clf)
```
