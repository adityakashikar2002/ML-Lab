```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('Iris.csv')
df
```

```python
X=df.drop(columns='Species',axis=1 )
```

```python
X
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=3,random_state=1)
km.fit(X)
```

```python
km.labels_
```

```python
km.cluster_centers_
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X['SL'],X['SW'],c=km.labels_)
plt.show()
```

```python
plt.scatter(X['PL'],X['PW'],c=km.labels_)
plt.show()
```

```python
inertia_scores=[]
for i in range(1,150):
    k=KMeans(n_clusters=i)
    k.fit(X)
    inertia_scores.append(k.inertia_)
```

```python
inertia_scores
```

# Till Above

```python
x_labels=range(1,150)
plt.plot(x_labels,inertia_scores,marker='o')
#plt.xlim(0,20)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
cluster=AgglomerativeClustering(n_clusters=3).fit(X)
```

```python
cluster
```

```python
cluster.labels_
```

```python
plt.scatter(X['SL'],X['SW'],c=cluster.labels_)
plt.show()
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(km.labels_,cluster.labels_)
```
