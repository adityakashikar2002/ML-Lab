# Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df1.head()
```

```python
encoder = LabelEncoder()
```

```python
new_car_data=pd.DataFrame()
new_car_data
```

```python
new_car_data['name'] = encoder.fit_transform(df1['name']) # converting car name to numeric data
```

```python
new_car_data
```

```python
new_car_data['year'] = df1['year']
new_car_data['miles'] = df1['miles']
new_car_data['condition'] = encoder.fit_transform(df1['condition']) # converting car condition to numeric data
new_car_data['price'] = df1['price'] 
```

```python
new_car_data
```

```python
#Now all column data is in numner now we can handle missing data.
```

```python
from sklearn.impute import SimpleImputer
import numpy as np
```

```python
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
```

```python
car_NaN_handle = imputer.fit_transform(new_car_data) 
```
